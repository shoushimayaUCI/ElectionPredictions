```python
import pandas as pd
import numpy as np
```

```python
file = "C:\\Users\\Roy\\Desktop\\Project\\CountyPopData\\RaceSexPopulations\\20102020RacePopulationbyCounty.csv"
```

```python
#Start with 2010-2020 dataset because it needs the most work
rs2020 = pd.read_csv(file,encoding = "ISO-8859-1")
```

```python
#Only want totals regardless of age
rs2020 = rs2020[rs2020["AGEGRP"] == 0]
```

```python
#Fix year mappings
yearMap = {1:0, 2:0, 3:2010, 4:2011, 5:2012, 6:2013, 7:2014,
           8:2015, 9:2016, 10:2017, 11:2018, 12:2019, 13:0,
           14:2020}

yearCol = rs2020['YEAR'].to_list()

for i in range(0,len(yearCol)):
    yearCol[i] = yearMap.get(yearCol[i])

rs2020['YEAR'] = yearCol
rs2020 = rs2020[rs2020["YEAR"] != 0]
```

```python
rs2020 = rs2020.drop(['SUMLEV','AGEGRP'], axis = 1)
rs2020.columns = rs2020.columns.str.lower()
```

```python
#Columns common to all datasets from 1990-2020 that we're interested in
rsCols = ["stname","ctyname","state","county","year",
          "tot_pop","tot_male","tot_female","nh_male",
          "nh_female","h_male","h_female","wa_male",
          "wa_female","ba_male","ba_female","ia_male",
          "ia_female","aa_male","aa_female","nhwa_male",
          "nhwa_female","nhba_male","nhba_female",
          "nhia_male","nhia_female","nhaa_male","nhaa_female",
          "hwa_male","hwa_female","hba_male","hba_female",
          "hia_male","hia_female","haa_male","haa_female"]
```

```python
rs2020 = rs2020[rsCols]
#2010-2020 done with it's unique wrangling needs, now add other years
#and finish wrangling common to all files
```

```python
file2 = "C:\\Users\\Roy\\Desktop\\Project\\CountyPopData\\RaceSexPopulations\\19901999RacePopulationbyCounty.csv"
file3 = "C:\\Users\\Roy\\Desktop\\Project\\CountyPopData\\RaceSexPopulations\\20002010RacePopulationbyCounty.csv"

rs1990 = pd.read_csv(file2,encoding = "ISO-8859-1")
rs2000 = pd.read_csv(file3,encoding = "ISO-8859-1")
```

```python
rs1990 = rs1990[rsCols]
rs2000 = rs2000[rsCols]
```

```python
#For some weird reasons some records repeat themselves in the 1990-1999 datafile
rs1990 = rs1990.drop_duplicates()

#Remove 2010 in 2000-2010 dataset because we have updated 2010 from 2010-2020 dataset
rs2000 = rs2000[rs2000["year"] != 2010]
```

```python
#concat the different years, fix ctyname column and calculate proportions instead of counts
raceSexPop = pd.concat([rs1990, rs2000, rs2020], ignore_index = True)
raceSexPop["ctyname"] = raceSexPop["ctyname"].str.replace(" County","")
```

```python
#Cast columns to float from object dtype
raceSexPop = raceSexPop.astype({'tot_pop':'float', 'tot_male':'float', 'tot_female':'float', 'nh_male':'float',
                    'nh_female':'float', 'h_male':'float', 'h_female':'float', 'wa_male':'float',
                    'wa_female':'float', 'ba_male':'float', 'ba_female':'float', 'ia_male':'float',
                    'ia_female':'float', 'aa_male':'float', 'aa_female':'float', 'nhwa_male':'float',
                    'nhwa_female':'float', 'nhba_male':'float', 'nhba_female':'float', 'nhia_male':'float',
                    'nhia_female':'float', 'nhaa_male':'float', 'nhaa_female':'float', 'hwa_male':'float',
                    'hwa_female':'float', 'hba_male':'float', 'hba_female':'float', 'hia_male':'float',
                    'hia_female':'float', 'haa_male':'float', 'haa_female':'float'})
```

```python
popCols = ["tot_male","tot_female","nh_male",
          "nh_female","h_male","h_female","wa_male",
          "wa_female","ba_male","ba_female","ia_male",
          "ia_female","aa_male","aa_female","nhwa_male",
          "nhwa_female","nhba_male","nhba_female",
          "nhia_male","nhia_female","nhaa_male","nhaa_female",
          "hwa_male","hwa_female","hba_male","hba_female",
          "hia_male","hia_female","haa_male","haa_female"]

# We want population proportions in each county, not counts
for colName in popCols:
    raceSexPop[colName] = raceSexPop[colName] / raceSexPop['tot_pop']
```

```python
raceSexPop = raceSexPop.astype({'tot_pop':'int'})
```
