```python
import pandas as pd
import numpy as np
import itertools
```

```python
education = pd.read_csv("C:\\Users\\Roy\\Desktop\\Project\\CountyPopData\\Education\\RawData\\Education.csv",
                       usecols=[0,1,2,27,28,29,30,35,36,37,38,43,44,45,46])
```

```python
education.columns = ['FIPS Code', 'State', 'County', '%_less_HS 1990', '%_HS_Only 1990',
                    '%_Some_College 1990', '%_Bach 1990','%_less_HS 2000', '%_HS_Only 2000',
                    '%_Some_College 2000', '%_Bach 2000','%_less_HS 2020', '%_HS_Only 2020',
                    '%_Some_College 2020', '%_Bach 2020']
```

```python
#Remove State/National/PuertoRico Totals
education = education[education["County"].str.contains("County")]
```

```python
education["County"] = education["County"].str.replace(" County", "")
```

```python
education
```

```python
counties = list(education['FIPS Code'])
years = list(range(1990,2021))
```

```python
countyYearCombo = list()

for i in itertools.product(counties, years):
    countyYearCombo.append(list(i))
```

```python
# THIS DICTIONARY CODE OBTAINED FROM 
# https://code.activestate.com/recipes/577305-python-dictionary-of-us-states-and-territories/
states = {
        'AK': 'Alaska',
        'AL': 'Alabama',
        'AR': 'Arkansas',
        'AS': 'American Samoa',
        'AZ': 'Arizona',
        'CA': 'California',
        'CO': 'Colorado',
        'CT': 'Connecticut',
        'DC': 'District of Columbia',
        'DE': 'Delaware',
        'FL': 'Florida',
        'GA': 'Georgia',
        'GU': 'Guam',
        'HI': 'Hawaii',
        'IA': 'Iowa',
        'ID': 'Idaho',
        'IL': 'Illinois',
        'IN': 'Indiana',
        'KS': 'Kansas',
        'KY': 'Kentucky',
        'LA': 'Louisiana',
        'MA': 'Massachusetts',
        'MD': 'Maryland',
        'ME': 'Maine',
        'MI': 'Michigan',
        'MN': 'Minnesota',
        'MO': 'Missouri',
        'MP': 'Northern Mariana Islands',
        'MS': 'Mississippi',
        'MT': 'Montana',
        'NA': 'National',
        'NC': 'North Carolina',
        'ND': 'North Dakota',
        'NE': 'Nebraska',
        'NH': 'New Hampshire',
        'NJ': 'New Jersey',
        'NM': 'New Mexico',
        'NV': 'Nevada',
        'NY': 'New York',
        'OH': 'Ohio',
        'OK': 'Oklahoma',
        'OR': 'Oregon',
        'PA': 'Pennsylvania',
        'PR': 'Puerto Rico',
        'RI': 'Rhode Island',
        'SC': 'South Carolina',
        'SD': 'South Dakota',
        'TN': 'Tennessee',
        'TX': 'Texas',
        'UT': 'Utah',
        'VA': 'Virginia',
        'VI': 'Virgin Islands',
        'VT': 'Vermont',
        'WA': 'Washington',
        'WI': 'Wisconsin',
        'WV': 'West Virginia',
        'WY': 'Wyoming'
}

for i in range(0, len(countyYearCombo)):
    countyName = education[education['FIPS Code'] == countyYearCombo[i][0]].iloc[0]["County"]
    stateAb = education[education['FIPS Code'] == countyYearCombo[i][0]].iloc[0]["State"]
    countyYearCombo[i].insert(1, countyName)
    countyYearCombo[i].insert(1, states.get(stateAb))
    
```

```python
countyYearCombo
```

```python
def interpolateAnnualVals (education, countyYearCombo, category):
    '''Estimate annual values for different educational attainment categories'''
    
    #Add the estimated %s for category
    for i in range(0, len(countyYearCombo)):
    
        currYr = countyYearCombo[i][3]
        currCounty = countyYearCombo[i][0] #FIPS Code
    
        if currYr < 2001: #Estimate using 1990-2000 data
        
            lower_edu_per = education[education['FIPS Code'] == currCounty].iloc[0][category + ' 1990']
        
            upper_edu_per = education[education['FIPS Code'] == currCounty].iloc[0][category + ' 2000']
        
            deltaYrs = currYr - 1990
            annual_change_est = (upper_edu_per - lower_edu_per) / 10
            
            estimateYr = lower_edu_per + (deltaYrs * annual_change_est)
            countyYearCombo[i].append(round(estimateYr,2))
    
        else: #Estimate using 2000-2020 data
        
            lower_edu_per = education[education['FIPS Code'] == currCounty].iloc[0][category + ' 2000']
        
            upper_edu_per = education[education['FIPS Code'] == currCounty].iloc[0][category + ' 2020']
        
            deltaYrs = currYr - 2000
            annual_change_est = (upper_edu_per - lower_edu_per) / 20
        
            estimateYr = lower_edu_per + (deltaYrs * annual_change_est)
            countyYearCombo[i].append(round(estimateYr,2))
    
```

```python
interpolateAnnualVals(education, countyYearCombo, '%_less_HS')
interpolateAnnualVals(education, countyYearCombo, '%_HS_Only')
interpolateAnnualVals(education, countyYearCombo, '%_Some_College')
interpolateAnnualVals(education, countyYearCombo, '%_Bach')
```

```python
edu = pd.DataFrame(countyYearCombo, columns = [
    "FIPS Code", "State", "County", "Year",
    "%_less_HS", "%_HS_Only", "%_Some_College", "%_Bachelors"
])
```

```python
#Split state and county fips code into own columns and make sure they're type is int
#If fips column is 4 digits only first digit is state fips last 3 are county
#If fips column is 5 digits first 2 digits are state fips last 3 are county
edu["FIPS Code"] = edu["FIPS Code"].astype(str)

edu["State FIPS Code"] = edu["FIPS Code"].str[:-3]
edu["County FIPS Code"] = edu["FIPS Code"].str[-3:]


edu = edu.astype({"FIPS Code":"int", "State FIPS Code":"int", "County FIPS Code":"int"})
```
